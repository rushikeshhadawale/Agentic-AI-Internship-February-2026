```python
#Problem 1: Smart Light Controller

class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.is_on = False  # Default state is OFF

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def display_status(self):
        if self.is_on:
            print(f"{self.light_name} is ON")
        else:
            print(f"{self.light_name} is OFF")


# Sample Input
light = SmartLight("Bedroom Light")
action = "ON"

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

light.display_status()
```

```python
#Problem 2: Employee ID Card System
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("----- Employee ID Card -----")
        print(f"Employee Name : {self.name}")
        print(f"Employee ID   : {self.emp_id}")
        print(f"Department    : {self.department}")
        print("----------------------------")


# Sample Input
employee = Employee("Rahul", "EMP102", "IT")

# Display ID Card
employee.display_id_card()
```

```python
#Problem 3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def save_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")


# Sample Input
contact = Contact("Anita", "9876543210")

# Display Output
contact.save_contact()
```

```python
#Problem 4: Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


# Sample Input
product = Product("Headphones", 2499)

# Display Output
product.print_price_tag()
```

```python
#Task 2:Build Advanced Preprocessing Function

#Implentation

import re

def preprocess_text(text):
    
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. Handle repeated characters (e.g., sooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 6. Remove special characters (keep only alphabets and spaces)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 8. Remove very short tokens (length ≤ 2)
    tokens = text.split()
    filtered_tokens = []
    
    for word in tokens:
        if len(word) > 2 or word in ["no", "not"]:
            filtered_tokens.append(word)
    
    return filtered_tokens
```

```python
#Task 3: Stress Testing

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for i, text in enumerate(test_sentences, 1):
    tokens = preprocess_text(text)
    cleaned_sentence = " ".join(tokens)

    print(f"\nExample {i}")
    print("Original Text   :", text)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
```

```python
#Task 4: Token Analytics

def analyze_tokens(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    return total_tokens, unique_tokens, round(avg_length, 2)


for i, text in enumerate(test_sentences, 1):
    tokens = preprocess_text(text)
    cleaned_sentence = " ".join(tokens)

    total, unique, avg_len = analyze_tokens(tokens)

    print(f"\nExample {i}")
    print("Original Text   :", text)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("Total Tokens    :", total)
    print("Unique Tokens   :", unique)
    print("Avg Token Length:", avg_len)
```

```python
#Task 5:Frequency Analysis

from collections import Counter

all_tokens = []

# Collect tokens from all sentences
for text in test_sentences:
    tokens = preprocess_text(text)
    all_tokens.extend(tokens)

# Create frequency counter
word_freq = Counter(all_tokens)

# Top 10 most frequent words
top_10 = word_freq.most_common(10)

# Top 5 least frequent words
least_5 = sorted(word_freq.items(), key=lambda x: x[1])[:5]

print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word}: {count}")

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word}: {count}")
```

```python
#Task 6: Build Full Pipeline

#Implementation

def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens = preprocess_text(text)   # use your Task 2 function
        
        all_tokens.extend(tokens)        # collect all tokens
        clean_sentences.append(" ".join(tokens))  # cleaned sentence
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
#Task 7: Error Handling

import re

def preprocess_text(text):
    
    # ✅ Handle None or non-string input
    if not isinstance(text, str):
        return []
    
    # ✅ Handle empty string
    if text.strip() == "":
        return []
    
    # 1. Lowercase
    text = text.lower()
    
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. Remove emojis (simple range)
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    
    # 6. Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 7. Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 8. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # ✅ If nothing remains after cleaning
    if text == "":
        return []
    
    # 9. Token filtering
    tokens = text.split()
    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]
    
    return filtered_tokens
```

```python
def full_pipeline(text_list):
    
    # ✅ Handle invalid input
    if not isinstance(text_list, list):
        return {"tokens": [], "clean_sentences": []}
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens) if tokens else "")
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
edge_cases = [
    "",                     # Empty string
    "😍😍🔥🔥",             # Only emojis
    "1234567890",           # Only numbers
    "!!!@@@",               # Only symbols
]

result = full_pipeline(edge_cases)

print(result)
```
